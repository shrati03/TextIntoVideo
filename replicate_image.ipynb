```python
directions = {
    "moving left": 1,
    "moving right": 2,
    "moving up": 3,
    "moving down": 4,
    "moving forward": 5,
    "moving backward": 6,
    "moving up-left": 7,
    "moving up-right": 8,
    "moving down-left": 9,
    "moving down-right": 10,
    "moving left-left": 11,
    "moving right-right": 12,
}
```

```python
len_words = [len(sentence.split()) for sentence in dialogue["Phrase"]]
second = 2
```

```python
frame_persistance_time = [(len_word / second) for len_word in len_words]
```

```python
fps = 24
num_frames = [
    int(frame_persistance_time[i] * fps) for i in range(len(frame_persistance_time))
]
(num_frames)
```

```python
sum(num_frames)
```

```python
num_frames_per_direction = [num_frame / len(directions) for num_frame in num_frames]
```

```python
refresh_directions = [
    "moving left-left",
    "moving left",
    "moving up-left",
    "moving up",
    "moving forward",
    "moving up-right",
    "moving right",
    "moving right-right",
    "moving down-right",
    "moving down",
    "moving backward",
    "moving down-left",
]
```

```python
new_image_files = []
```

```python
k = 0
for i in range(dialogue.shape[0]):
    for j in range(int(num_frames_per_direction[k])):
        a = (-1) ** j
        refresh_directions = refresh_directions[::a]
        for direction in refresh_directions:
            og_image = Image.open(f"Frames\Frame1\{i+1}_{directions[direction]}.png")
            replicated_image = og_image.copy()
            replicated_image.save(
                f"Frames\Frame2\{i+1}_{j+1}_{directions[direction]}.png"
            )
    k += 1
```
